# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets

import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Load

```python
iris = datasets.load_iris()
x = iris.data
y = iris.target
```

```python

#Input data 

data=pd.DataFrame(iris['data'],columns=["Petal length","Petal Width","Sepal Length","Sepal Width"])
data['Species']=iris['target']
data['Species']=data['Species'].apply(lambda x: iris['target_names'][x])
```

```python
data.head()
```

```python
print(data.info())
```

```python
data.describe()
```

# Data Visualization

```python
plt.figure(figsize = (20,20))
sns.heatmap(data.corr(), annot= True , cmap='coolwarm', linewidths=2)
```

```python
sns.pairplot(data)
```

```python
sns.countplot(data['Species'])
```

```python
# Scatter plot of data based on Sepal Length and Width features
sns.FacetGrid(data,hue='Species').map(plt.scatter,'Sepal Length','Sepal Width').add_legend()
plt.show()

# Scatter plot of data based on Petal Length and Width features
sns.FacetGrid(data,hue='Species').map(plt.scatter,'Petal length','Petal Width').add_legend()
plt.show()
```

# Train Test Slit

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=1) 
```

# Model Building

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import sklearn.metrics as sm

tree_classifier = DecisionTreeClassifier()
tree_classifier.fit(x_train,y_train)
y_pred = tree_classifier.predict(x_test)
sm.accuracy_score(y_test, y_pred)
```

# Comparing the actual and predicted flower classification

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}) 
df
```
