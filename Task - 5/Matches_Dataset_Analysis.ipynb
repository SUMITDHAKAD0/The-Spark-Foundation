```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Data load

```python
data = pd.read_csv('matches.csv')
print('data loaded')
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data = data.drop(['umpire3'], axis=1)
```

```python
data.info()
```

Fill Missing Values

```python
#only int and float value are handle
median = data.median()
#median
data = data.fillna(median)
data.info()
```

Total Season played

```python
len(data['season'].unique())
```

```python
data['season'].unique()
```

Total Teams

```python
pd.concat([data['team1'],data['team2']]).unique()
```

```python
len(pd.concat([data['team1'],data['team2']]).unique())
```

Winning Teams

```python
plt.figure(figsize = (10,5))
sns.countplot(x = 'winner', data = data)
plt.title('Winning Teams')
plt.xticks(rotation = 90);
```

Win by Runs

```python
win_max=data.groupby(['winner'],as_index=False)['win_by_runs'].max()
plt.figure(figsize = (10,5))
sns.barplot(win_max['win_by_runs'], win_max['winner'])
plt.title('Winning Teams')
plt.xticks(rotation = 180);
```

```python

win_max=data.groupby(['winner'],as_index=False)['win_by_runs'].max()
plt.subplots(figsize=(10,5))
plt.xticks(rotation=90)
sns.barplot(win_max['winner'],win_max['win_by_runs'])
```

Man Of The Matches

```python
data['player_of_match'].value_counts()[0:10]
#data['player_of_match'].value_counts().head(10)
```

```python

plt.figure(figsize=(10,5))
#sns.barplot(x=data['player_of_match'].value_counts(),y=data['player_of_match'].value_counts())
sns.barplot(x= data.player_of_match.value_counts()[:10].index, y= data.player_of_match.value_counts()[:10])
plt.title('Man Of The Match Winner')
plt.xlabel('Player')
plt.ylabel('Match')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x= data.player_of_match.value_counts(), y= data.player_of_match.value_counts())
plt.title('Man of the match Winners')
plt.xlabel('Player')
plt.ylabel('Match')
plt.xticks(rotation = 90)
```

Number of Matches in each Season

```python
#sns.countplot(data['season'].sort_values())
plt.subplots(figsize=(10,5))
sns.countplot(x=data['season'].sort_values())
plt.title('Number of matches in each Season', fontsize = 15)
plt.show()
```

Toss Winners

```python
plt.subplots(figsize=(10,5))
sns.countplot(x=data['toss_winner'].sort_values())
plt.title('Toss winners')
plt.xticks(rotation = 90)
plt.show()
```

Comparinh the Toss Decisions

```python
plt.figure(figsize=(15,5))
sns.countplot(x=data['season'].sort_values(), hue = data['toss_decision'])
plt.title("Decision to field or bat across seasons")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

 When a Team wins by Runs

```python
runs = data[data['win_by_runs']!=0]
plt.figure(figsize=(10,10))
plt.pie(runs['winner'].value_counts(), labels=runs['winner'].value_counts().keys()
        ,autopct='%0.1f%%')
```

```python
wickets = data[data['win_by_wickets']!=0]
plt.figure(figsize=(10,10))
plt.pie(wickets['winner'].value_counts(), labels=wickets['winner'].value_counts().keys()
        ,autopct='%0.1f%%')
```

Match Vanues

```python
plt.subplots(figsize=(15,5))
sns.countplot(x ='venue',data = data,order = data['venue'].value_counts().index)
plt.title('Match Venues')
plt.xlabel('Stadium')
plt.ylabel('Matches')
plt.xticks(rotation = 90)
```

# Conclusion:
